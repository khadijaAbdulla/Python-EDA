```python
ls = []
shows=[]
hasMore = True
count = 1
querystring = {"series_granularity":"show",
                   "order_direction":"asc",
                   "order_by":"original_title",
                   "genres_relation":"and",
                   "output_language":"en",
                   "country" : 'us'
                  } # the filter which retreives serieses/movies
    
response = requests.get(url, headers=headers, params=querystring)
ls.append(response.json()) # get the result for the first request

ls2 = ls[0] # get the dictionary inside the result list (inc Shows, hasMore and nextCursor)
shows = ls2['shows'] #get the dictionary of shows which can be converted to a dataframe
hasMore = ls2['hasMore'] #get the value of hasMore (True or False)
nextCursor = ls2['nextCursor'] #get the cursor for the next page and add it to query the next page
print(hasMore,nextCursor)

# a loop to go through the next pages and send additional requests if more shows are available
while count < 9:
    print('inside the loop' + str(count))
    if hasMore == True:
        querystring = {"series_granularity":"show",
                       "order_direction":"asc",
                       "order_by":"original_title",
                       "genres_relation":"and",
                       "output_language":"en",
                       "country" : "us",
                       "cursor": nextCursor
                      } # the filter which retreives serieses/movies
        print('inside the if statement')
        response = requests.get(url, headers=headers, params=querystring)
        ls = response.json() # get the result for the other requests
        
        #ls2 = ls[0] # get the dictionary inside the result list (inc Shows, hasMore and nextCursor)
        shows.extend(ls['shows']) #get the dictionary of shows which can be converted to a dataframe
        hasMore = ls['hasMore'] #get the value of hasMore (True or False)
        nextCursor = ls['nextCursor'] #get the cursor for the next page and add it to query the next page
        print(hasMore,nextCursor)
        count+=1
    else:
        break
```
